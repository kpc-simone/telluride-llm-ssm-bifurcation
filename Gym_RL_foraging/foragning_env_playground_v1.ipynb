# Testing the foraging environment

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from collections import defaultdict, deque
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.animation as animation
import seaborn as sns
import gym
from gym import spaces
from IPython.display import clear_output, display, HTML



from ForagingEnvNeuromod import *

# set matplotlib font to arial
plt.rcParams['font.family'] = 'Arial'

%load_ext autoreload
%autoreload 2
```

```python
# Create the environment
env = ForagingEnvNeuromod()
env.reset()

num_frames = 100
# Generate frames and save them as PNG files
for i in range(num_frames):
    action = np.random.choice(env.action_space.n)  # Select a random action for demonstration
    env.step(action)
    env.render(num_frames)

# Create the animation using the saved PNG files
fig = plt.figure()
frames = []


for i in range(num_frames):
    img = plt.imread(f'frames/frame_{i:04d}.png')
    frame = plt.imshow(img, animated=True)
    frames.append([frame])

ani = animation.ArtistAnimation(fig, frames, interval=200, repeat=False)

# Save animation as GIF
ani.save('foraging_animation.gif', writer='Pillow', fps=10)

# Delete all PNG files after creating the animation
for i in range(num_frames):
    os.remove(f'frames/frame_{i:04d}.png')
```
