```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
nln_df = pd.read_csv('expt_tstart10_tend20_taulearn0.02_winsize10.csv')
lrn_df = pd.read_csv('expt_tstart0_tend20_taulearn0.02_winsize10.csv')

fig,(ax1,ax2) = plt.subplots(2,1,sharex=True,sharey=True)
ax1.plot(nln_df['time'],nln_df['error'],color='k')

subdf = nln_df[nln_df['time']>20]
mse_nolearn = np.linalg.norm(subdf['error'])
print('no learning: ', mse_nolearn)

ax2.plot(lrn_df['time'],lrn_df['error'],color='k')
subdf = lrn_df[lrn_df['time']>20]
mse_learn = np.linalg.norm(subdf['error'])
print('learning: ', mse_learn)
      
plt.show()
```
