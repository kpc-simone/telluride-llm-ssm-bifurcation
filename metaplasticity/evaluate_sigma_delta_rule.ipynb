```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
nln_df = pd.read_csv('expt_tstart10_tend20_taulearn0.05_winsize100.csv')
lrn_df = pd.read_csv('expt_tstart0_tend20_taulearn0.05_winsize100.csv')
bsl_df = pd.read_csv('pes_baseline.csv')

fig,(ax1,ax2) = plt.subplots(2,1,sharex=True,sharey=True)
#ax1.plot(nln_df['time'],nln_df['error'],color='k')
#subdf = nln_df[nln_df['time']>20]
#mse_nolearn = np.linalg.norm(subdf['error'])
#print('no learning: ', mse_nolearn)

ax1.plot(bsl_df['time'],bsl_df['error'],color='k')
subdf = bsl_df[bsl_df['time']>20]
mse_bsl = np.linalg.norm(subdf['error'])
title = r'MSE, $\Delta$ Rule: ' + f'{round(mse_bsl,2)}'
ax1.set_title(title)
    
ax2.plot(lrn_df['time'],lrn_df['error'],color='k')
subdf = lrn_df[lrn_df['time']>20]
mse_learn = np.linalg.norm(subdf['error'])
title = r'MSE, $\sigma-\Delta$ Rule: ' + f'{round(mse_learn,2)}'
ax2.set_title(title)

ax2.set_xlabel('Time(s)')

for ax in (ax1,ax2):
    ax.set_ylabel('Error')
    ax.axvspan(20,30,color='orangered',alpha=0.1)
    for spine in ['top','right']:
        ax.spines[spine].set_visible(False)

plt.show()
```
