```python
import matplotlib.pyplot as plt
from tqdm import tqdm
import pandas as pd
import numpy as np
import sys,os

import cupy as cp
import nengo

from sklearn.cluster import estimate_bandwidth
from scipy.special import betaincinv
from scipy.stats import beta
```

```python
def make_unitary_matrix_fourier( ssp_dim, domain_dim, eps=1e-3, rng = cp.random):
    a = rng.rand( (ssp_dim - 1)//2, domain_dim )
    sign = rng.choice((-1, +1), size=cp.shape(a) )
    phi = sign * cp.pi * (eps + a * (1 - 2 * eps))

    fv = cp.zeros( (ssp_dim,domain_dim), dtype='complex64')
    fv[0,:] = 1

    fv[1:(ssp_dim + 1) // 2,:] = phi
    fv[-1:ssp_dim // 2:-1,:] = -fv[1:(ssp_dim + 1) // 2,:]
    
    if ssp_dim % 2 == 0:
        fv[ssp_dim // 2,:] = 1

    return fv

class SSPEncoder:
    def __init__(self, phase_matrix, length_scale):
        '''
        Represents a domain using spatial semantic pointers.

        Parameters:
        -----------

        phase_matrix : cp.ndarray
            A ssp_dim x domain_dim ndarray representing the frequency 
            components of the SSP representation.

        length_scale : float or cp.ndarray
            Scales values before encoding.
        '''
        self.phase_matrix = phase_matrix
        self.domain_dim = self.phase_matrix.shape[1]
        self.ssp_dim = self.phase_matrix.shape[0]
        self.update_lengthscale(length_scale)

    def update_lengthscale(self, scale):
        '''
        Changes the lengthscale being used in the encoding.
        '''
        if not isinstance(scale, cp.ndarray) or scale.size == 1:
            self.length_scale = scale * cp.ones((self.domain_dim,))
        else:
            assert scale.size == self.domain_dim
            self.length_scale = scale
        assert self.length_scale.size == self.domain_dim
    
    def encode(self,x):
        '''
        Transforms input data into an SSP representation.

        Parameters:
        -----------
        x : cp.ndarray
            A (num_samples, domain_dim) array representing data to be encoded.

        Returns:
        --------
        data : cp.ndarray
            A (num_samples, ssp_dim) array of the ssp representation of the data
            
        '''
        
        x = cp.atleast_2d(x)
        ls_mat = cp.atleast_2d(cp.diag(1/self.length_scale.flatten()))
        
        assert ls_mat.shape == (self.domain_dim, self.domain_dim), f'Expected Len Scale mat with dimensions {(self.domain_dim, self.domain_dim)}, got {ls_mat.shape}'
        scaled_x = x @ ls_mat
        data = cp.fft.ifft( cp.exp( 1.j * self.phase_matrix @ scaled_x.T), axis=0 ).real
        
        return data.T   
    
def RandomSSPSpace(domain_dim, ssp_dim, length_scale = None, 
                   rng = cp.random.default_rng() ):
    
    phase_matrix = make_unitary_matrix_fourier(ssp_dim,domain_dim)

    length_scale = cp.array( length_scale )
    return SSPEncoder(phase_matrix, length_scale=length_scale)

def bind(a, b):
    '''
    Binds (circular convolution) two sets of vectors together
    '''

    #assert a.shape[1] == b.shape[1], f'Expected SSPs to have same dimensionality.  Got {self.shape} * {b.shape}'
    a = cp.atleast_2d(a)
    b = cp.atleast_2d(b)

    return cp.fft.ifft( cp.multiply(cp.fft.fft(a, axis=1), cp.fft.fft(b,axis=1)) , axis=1).real#.view(SSP)

def invert(a):
    '''
    Implements the pseudo-inverse of the SSP
    '''
    return a[:,-cp.arange(a.shape[1])]
```

```python
# "environment"

# toy model world
class Env(object):
    def __init__(self):
        self.n_actions = 4
        self.directions = cp.array(
                        [[0.],[0.5*cp.pi],[cp.pi],[1.5 * cp.pi]]
                        )
        self.state = cp.array([[0.,0.]])
    def step(self,action = None):
        if action is not None:
            dx = cp.cos(self.directions[action])
            dy = cp.sin(self.directions[action])
            self.state += cp.hstack( [dx,dy] )

    def reset(self):
        self.state = cp.array([[0.,0.]])        
    
env = Env()
n_syllables = env.n_actions
print(env.directions)
```

```python
# will need to bind the state with the current option to decode termination probability; otherwise skills cannot be reused
# policy over options is a function of the state only as per option-critic

# first step. Can I follow one sequence?
# Toy problem: low-level actions available to the agent include N,W,S,E

# using Bob Datta's terminology, sequences can be up to 3 syllables long.
# I'll follow it for 13 timesteps

# first, generate the syllabary (the set of action primitives available to the agent. 
# In a discrete space this is just the possible actions. In continuous space this would be more complex. 

# vocabulary creation

action_dim = 1
time_dim = 1
ssp_dim = 1024
act_space = RandomSSPSpace( domain_dim = action_dim, ssp_dim = ssp_dim, length_scale = 1. )
step_space = RandomSSPSpace( domain_dim = time_dim, ssp_dim = ssp_dim, length_scale = 1. )

syl_phis = cp.zeros( (env.directions.shape[0],ssp_dim) )
for idx,syl in enumerate(env.directions):
    syl_phis[idx,:] = act_space.encode(syl.reshape(-1,action_dim))

seq_len = 4
step_phis = cp.zeros( ( seq_len, ssp_dim ) )
for step in range( seq_len ):
    step_phis[step,:] = step_space.encode( step )

# 0: east
# 1: north
# 2: west
# 4: south

def gen_sequence( syl_phis, step_phis ):
    seq_phi = cp.zeros( (1,ssp_dim) )
    for idx in range( syl_phis.shape[0] ):
        seq_phi += bind( cp.atleast_2d(syl_phis[idx,:]), cp.atleast_2d(step_phis[idx,:]) )
    return seq_phi
seq_phis = cp.vstack([syl_phis[0,:],syl_phis[1,:],syl_phis[1,:]])
print(seq_phis.shape)
seq_phi = gen_sequence( seq_phis, step_phis )
```

```python
# what is at the last position?
query_v = invert( cp.atleast_2d(step_phis[0,:]) )
query_r = bind( seq_phi, query_v )

# equivalence between binding and holographic recurrent networks?
# check that each is dissimilar from the other
# in continuous action spaces, could leverage decoding noise to let the intraoption policies mutate/drift

fig,ax = plt.subplots(1,1)
sims = cp.dot(query_r,syl_phis.T).get()
ax.plot(sims.flatten())
ax.axvline(np.argmax(sims),color='k',linestyle='--')
plt.show()
```

```python
timesteps = 13
t_period = 3

env.reset()
states = cp.zeros( (13,2) )
# limit cycle attractor w/ SSPs?
for t in range(timesteps):
    t_idx = t % t_period

    query_v = invert( cp.atleast_2d(step_phis[t_idx,:]) )
    query_r = bind( seq_phi, query_v )
    sims = cp.dot(query_r,syl_phis.T)
    action = cp.argmax( sims )

    env.step( action )
    states[t,:] = env.state

states = states.get()
fig,ax = plt.subplots(1,1)
ax.plot(states[:,0],states[:,1])
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
plt.show()
```

```python
def gen_random_sequences( syl_phis, n_sequences = 100, n_syllables_max = 21 ):

    seq_phis = cp.zeros( (n_sequences,ssp_dim) )
    
    step_phis = cp.zeros( ( n_syllables_max, ssp_dim ) )
    for step in range( n_syllables_max):
        step_phis[step,:] = step_space.encode( step )
        
    for v in range( n_sequences ):
        n_syls = np.random.choice( range(1,n_syllables_max+1) )
        for t in range( n_syls ):
            idx = np.random.choice( range(n_syllables) )

            seq_phis[v,:] += bind( cp.atleast_2d(syl_phis[idx,:]), cp.atleast_2d(step_phis[t,:]) ).flatten()
            
    return seq_phis, step_phis

seq_phis, step_phis = gen_random_sequences( syl_phis )
```

```python
for s in range(seq_phis.shape[0]):
    seq_phi = seq_phis[s,:]

    t_period = 21#13#n_syllables_max
    timesteps = 20#n_syllables_max
    
    env.reset()
    states = cp.zeros( (timesteps,2) )
    # limit cycle attractor w/ SSPs?
    for t in range(timesteps):
        t_idx = t % t_period
    
        query_v = invert( cp.atleast_2d(step_phis[t_idx,:]) )
        query_r = bind( seq_phi, query_v )
        sims = cp.dot(query_r,syl_phis.T)
        if cp.max(sims) > 0.5:
            action = cp.argmax( sims )
        else:
            action = None
        env.step( action )
        states[t,:] = env.state
    
    states = states.get()
    fig,ax = plt.subplots(1,1)
    ax.plot(states[:,0],states[:,1])
    ax.set_xlim(-10,10)
    ax.set_ylim(-10,10)
    plt.show()
```
